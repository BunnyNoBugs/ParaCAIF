```python
import pandas as pd
import seaborn as sns
```

```python
alpha_df = pd.DataFrame()
alpha_df['alpha'] = [x for x in [-5, -3, -2, -1, 0] for i in range(4)]
alpha_df['metric'] = ['ACC', 'SIM', 'FL', 'J'] * 5
alpha_df['score'] = [0.68, 0.65, 0.69, 0.30] + [0.69, 0.66, 0.70, 0.31] + [
    0.70, 0.65, 0.69, 0.31] + [0.68, 0.63, 0.70, 0.29] + [0.61, 0.60, 0.71, 0.25]
```

```python
ax = sns.lineplot(x='alpha', y='score', hue='metric', style='metric', data=alpha_df,
                  markers=True, dashes=False)
fig = ax.get_figure()
fig.savefig('img/alpha.pdf')
```
