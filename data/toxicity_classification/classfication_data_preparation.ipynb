```python
import pandas as pd
```

```python
!wget https://raw.githubusercontent.com/skoltech-nlp/russe_detox_2022/main/data/input/train.tsv
!wget https://raw.githubusercontent.com/skoltech-nlp/russe_detox_2022/main/data/input/dev.tsv
```

```python
train = pd.read_csv("train.tsv", sep="\t")
train = train.drop(columns=['index'])
dev = pd.read_csv("dev.tsv", sep="\t")
train
```

```python
train_df = pd.DataFrame()
train_df['text'] = pd.concat([train['toxic_comment'], train['neutral_comment1']])
train_df['is_toxic'] = pd.concat([pd.Series([1] * len(train)), pd.Series([0] * len(train))])
train_df = train_df.reset_index(drop=True)
train_df
```

```python
dev_df = pd.DataFrame()
dev_df['text'] = pd.concat([dev['toxic_comment'], dev['neutral_comment1']])
dev_df['is_toxic'] = pd.concat([pd.Series([1] * len(dev)), pd.Series([0] * len(dev))])
dev_df = dev_df.reset_index(drop=True)
dev_df
```

```python
train_df.to_csv('classification_train.csv', index=False)
dev_df.to_csv('classification_dev.csv', index=False)
```
